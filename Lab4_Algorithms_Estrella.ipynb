```python
# Step 1: Define an algorithm in pseudocode as comments
# Algorithm to find the largest number in a list
# 1. Assume the first number is the largest
# 2. Compare each number with the current largest
# 3. If a number is larger, update the largest
# 4. Return the largest number
# Step 2: Implement it in Python

numbers = [10, 5, 20, 8, 15] 
largest = numbers[0] 
for num in numbers: 
    if num > largest: 
        largest = num 

print("Largest number:", largest) 
```

```python
# Step 1: Implement linear search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
# Step 2: Test the function
data = [3, 5, 2, 4, 9]
index = linear_search(data, 4)
print("Found at index:", index)
```

```python
# Step 1: Implement binary search
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
# Step 2: Test with a sorted list
sorted_data = [1, 3, 5, 7, 9, 11]
result = binary_search(sorted_data, 7)
print("Found at index:", result)
```

```python
# Step 1: Implement bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
             arr[j], arr[j + 1] = arr[j + 1], arr[j]
# Step 2: Test sorting
to_sort = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(to_sort)
print("Sorted list:", to_sort)
```

```python
# Step 1: Implement selection sort
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
# Step 2: Test and observe performance
sample_data = [29, 10, 14, 37, 13]
selection_sort(sample_data)
print("Sorted list:", sample_data)
```

```python
# Step 1: Implement insertion sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
# Step 2: Test and measure performance
arr = [12, 11, 13, 5, 6]
insertion_sort(arr)
print("Sorted array:", arr)
```

```python
def fib_iter(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
print(fib_iter(10))
```

```python
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n-1) + fib_rec(n-2)
print(fib_rec(10))
```

```python
import time
import random

def bubble_sort(arr):
    n = len(arr)
    swaps = comparisons = 0
    for i in range(n):
        for j in range(0, n - i - 1):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
    return comparisons, swaps

def selection_sort(arr):
    n = len(arr)
    swaps = comparisons = 0
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        swaps += 1
    return comparisons, swaps

def insertion_sort(arr):
    n = len(arr)
    swaps = comparisons = 0
    for i in range(1, n):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            comparisons += 1
            arr[j + 1] = arr[j]
            swaps += 1
            j -= 1
        arr[j + 1] = key
    return comparisons, swaps

student_id = 20250590
list_size = (student_id % 1000) + 50
original_list = random.sample(range(1, 2000), list_size)

results = []
for name, func in [("Bubble Sort", bubble_sort), 
                  ("Selection Sort", selection_sort), 
                  ("Insertion Sort", insertion_sort)]:
    data_copy = original_list.copy()
    start = time.time()
    comps, swaps = func(data_copy)
    duration = time.time() - start
    results.append([name, comps, swaps, f"{duration:.6f}"])
print(f"{'Algorithm':<15} | {'Comps':<10} | {'Swaps':<10} | {'Time (s)':<10}")
print("-" * 55)
for r in results:
    print(f"{r[0]:<15} | {r[1]:<10} | {r[2]:<10} | {r[3]:<10}")
```
