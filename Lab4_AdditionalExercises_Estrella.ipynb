```python
import time

def profile_algorithms():
    # Input Data
    student_id = "TUPM-25-0590"
    target = int(student_id[-2:])

    search_list = list(range(1000000))
    duplicate_test_list = list(range(5000)) + [99]

    # 1. Linear Search O(n)
    start = time.time()
    for i in search_list:
        if i == target:
            break
    linear_time = time.time() - start

    # 2. Binary Search O(log n)
    start = time.time()
    low, high = 0, len(search_list) - 1
    while low <= high:
        mid = (low + high) // 2
        if search_list[mid] == target:
            break
        elif search_list[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    binary_time = time.time() - start

    # 3. Quadratic Search O(n^2)
    start = time.time()
    found_dup = False
    for i in range(len(duplicate_test_list)):
        for j in range(i + 1, len(duplicate_test_list)):
            if duplicate_test_list[i] == duplicate_test_list[j]:
                found_dup = True
                break
        if found_dup:
            break
    quadratic_time = time.time() - start

    # Output
    print(f"{'Algorithm':<20} | {'Time Taken (sec)':<20}")
    print("-" * 45)
    print(f"{'Linear Search':<20} | {linear_time:.10f}")
    print(f"{'Binary Search':<20} | {binary_time:.10f}")
    print(f"{'Nested Loop (Dup)':<20} | {quadratic_time:.10f}")


profile_algorithms()
```

```python
import time

def iterative_fibonacci_analysis():
    student_id = "TUPM-25-0590"  # Replace with your actual student_id
    
    # Get last digit of student ID
    n = (int(student_id[-1]) % 10) + 15

    start = time.time()

    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b

    end = time.time()

    print(f"{'Problem Size (n)':<20} | {'Time Taken':<15} | {'Result':<10}")
    print("-" * 55)
    print(f"{n:<20} | {end - start:.10f} | {a:<10}")
    print("\nAnalysis: This iterative approach has O(n) linear complexity.")

iterative_fibonacci_analysis()
```

```python
import time
import random

def bubble_sort(arr):
    n = len(arr)
    swaps = comparisons = 0
    for i in range(n):
        for j in range(0, n - i - 1):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
    return comparisons, swaps

def selection_sort(arr):
    n = len(arr)
    swaps = comparisons = 0
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        swaps += 1
    return comparisons, swaps

def insertion_sort(arr):
    n = len(arr)
    swaps = comparisons = 0
    for i in range(1, n):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            comparisons += 1
            arr[j + 1] = arr[j]
            swaps += 1
            j -= 1
        arr[j + 1] = key
    return comparisons, swaps

student_id = 20250590
list_size = (student_id % 1000) + 50
original_list = random.sample(range(1, 2000), list_size)

results = []
for name, func in [("Bubble Sort", bubble_sort), 
                  ("Selection Sort", selection_sort), 
                  ("Insertion Sort", insertion_sort)]:
    data_copy = original_list.copy()
    start = time.time()
    comps, swaps = func(data_copy)
    duration = time.time() - start
    results.append([name, comps, swaps, f"{duration:.6f}"])
print(f"{'Algorithm':<15} | {'Comps':<10} | {'Swaps':<10} | {'Time (s)':<10}")
print("-" * 55)
for r in results:
    print(f"{r[0]:<15} | {r[1]:<10} | {r[2]:<10} | {r[3]:<10}")
```
